```python
import pandas as pd
import os

#define file paths to learning and testing data repositories
path = "C:/Users/danna/Documents/GitHub/WBBTournamentPredictions/MachineLearning/LearningData"
path2 = "C:/Users/danna/Documents/GitHub/WBBTournamentPredictions/MachineLearning/TestData"

#PRE: path is the file path to data to be manipulated, isLearning is a bool that determines if the given file is learning data
#POST: 
def featureManip(path, isLearning):
    #get items in the given location
    yr_list = os.listdir(path)
#     print(yr_list)
    gamesLists = []
    nonFolders = 0
    #take out non folders
    for yr in yr_list:
        try:
            path2 = path + "/" + yr
            gamesLists.append(os.listdir(path2))
        except:
            print('error, found a not-folder')
            nonFolders+=1
    
    #define universal names for files to be saved as
    team1File = "Team 1 Basic BS"
    team2File = "Team 2 Basic BS" 
    ext = ".csv"
    
    base_cols = None

    games = []
    gamesAnswers = []
    gameID = 0

    print("\n\nSTART")
    
    #go through each year folder
    for yr in range( len(yr_list)  - nonFolders):
        #go through each game file in each year
        for gameFile in range( len(gamesLists[yr])):
            #get file paths
            testGamePath = path + '/' + yr_list[yr] + '/' + gamesLists[yr][gameFile] 
            team1DataPath = testGamePath + '/' + team1File + ext
            team2DataPath = testGamePath + '/' + team2File + ext
            #try to read the file, prints error and continues if cannot open game
            try:
                team1 = pd.read_csv(team1DataPath)
                team2 = pd.read_csv(team2DataPath)
            except: 
                print("error with", gamesLists[yr][gameFile], "(", yr_list[yr], ")... skipping.")
                continue

            teams = [team1, team2]
            summaries = []
            #go through both team 1 and team 2
            for team in teams:
                #drop index column and rename team column
                team.drop(["Unnamed: 0"], axis=1, inplace=True)
                team.iloc[-1, 0] = team.columns[0]
                team.rename(columns={team.columns[0] : "TEAM"}, inplace=True)    
                
                team["YEAR"] = yr_list[yr]
                
                #convert to percentages
                team["FG%"] = team["FG"] / team["FGA"]
                team["2P%"] = team["2P"] / team["2PA"]
                team["3P%"] = team["3P"] / team["3PA"]
                team["FT%"] = team["FT"] / team["FTA"]
                
                #fill any null cells with 0 (for cases of dividing by 0)
                team["FG%"] = team["FG%"].fillna(0)
                team["2P%"] = team["2P%"].fillna(0)
                team["3P%"] = team["3P%"].fillna(0)
                team["FT%"] = team["FT%"].fillna(0)
                
                #drop non-percentage columns
                team.drop(["FG", "FGA", "2P", "2PA", "3P", "3PA"], axis=1, inplace=True)
                
                #get the total of each stat for the team
                totals = pd.concat([team.iloc[-1, :]])
                totals.index = totals.index.str.upper()
                
                #get maximum of each stat over the remaining players
                maxes = pd.concat([team.iloc[:-1, :].max()])
                maxes.index = maxes.index.str.upper() + "_MAX"
                
                #connect totals and max stat value tables
                summary = pd.concat([totals, maxes])
                #add to game summaries array
                summaries.append(summary)
            
            #concatenate all summaries into rows in a dataframe
            summary = pd.concat(summaries, axis=1).T

            #mark team 2 stats as _OPP
            game = pd.concat(summaries, axis=1).T
            game_opp = game.iloc[::-1].reset_index()           
            game_opp.columns += "_OPP"
            
            #concatentate team 1 and team 2 stats in same row
            full_game = pd.concat([game, game_opp], axis=1)
            
            #drop unneeded columns
            droppedCols = ["YEAR_MAX", "YEAR_OPP", "YEAR_MAX_OPP", "TEAM_MAX", "TEAM_MAX_OPP", "index_OPP"]
            full_game.drop(droppedCols, axis=1, inplace=True)
            
            #determine winner of game
            full_game["WIN"] = full_game["PTS"] > full_game["PTS_OPP"]
                        
            full_game.drop(["PTS", "PTS_OPP"], axis=1, inplace=True)
            
            #make gameID for possible later use
            full_game["GAME_ID"] = gameID
            gameID += 1
            
            #handle whether the given data was for testing or learning
            #if for testing, do not include WIN column in same file
            ans = full_game[['WIN']].copy()
            if(not isLearning):
                full_game.drop(["WIN"], axis=1, inplace=True)

            games.append(full_game)
            gamesAnswers.append(ans)
            
            #print progess notifications
            if len(games) % 50 == 0:
                print(f"{len(games)} / {len(yr_list) * len(gamesLists)}")

    print("END")
    if(isLearning):
        return(games)
    else:
        return(games, gamesAnswers)
#line for learning data    
# games = featureManip(path, True)
#line for testing data
games, gamesAnswers = featureManip(path2, False)
```

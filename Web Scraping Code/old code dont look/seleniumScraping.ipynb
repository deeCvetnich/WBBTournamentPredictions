```python
#importing packages
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
url='https://www.sports-reference.com/cbb/seasons/'

#PRE: given specfic start url for the web database sports-reference, bool for using headless mode
#POST: returns arrays of tournament years, links to tournament pages
def getTourneyYearLinks(url, headlessMode):
    print("-> START 'getTourneyYearLinks'")

    DRIVER_PATH = '/path/to/chromedriver'

    print("headless mode: " + str(headlessMode))

    options = Options()
    #configure non/headless mode:
    options.headless = headlessMode
    #configure window size:
    #options.add_argument("--window-size=1920,1200")
    driver = webdriver.Chrome(options=options, executable_path=DRIVER_PATH)

    #open given url
    print("opening url: " + url + " ...")
    driver.get(url)
    print("url opened!")

    wait = WebDriverWait(driver, 30)

    #start automation
    print("getting tourneys...")
    driver.find_element(By.XPATH, '//*[@id="all_seasons_wrap"]/div[3]/div[2]/a').click()
    tourneys = driver.find_elements(By.XPATH, '//*[@id="seasons_NCAAW"]/tbody/tr[*]/td[2]/a')

    #check variable exists and is not empty
    if(tourneys != None or len(tourneys)!=0):
        print("tourneys links found!")
    else: 
        print("could not find tourneys")
        driver.quit()

    tourneyYears = []
    tourneyLinks = []
    for i in range(len(tourneys)):
        tourneyYears.append(tourneys[i].text)
        tourneyLinks.append(tourneys[i].get_attribute('href'))
    print("quitting driver")
    driver.quit()
    
    print("FINISH 'getTourneyYearLinks'")
    print()
    
    return tourneyYears, tourneyLinks
```

```python
#PRE: given url to game, headlessMode bool
#POST: returns array of links to all games in that tournament season
def getTourneyGameLinks(url, headlessMode):
    print("-> START 'getTourneyGameLinks'")
    
    DRIVER_PATH = '/path/to/chromedriver'

    print("headless mode: " + str(headlessMode))

    options = Options()
    #configure non/headless mode:
    options.headless = headlessMode
    #configure window size:
    #options.add_argument("--window-size=1920,1200")
    driver = webdriver.Chrome(options=options, executable_path=DRIVER_PATH)

    #open given url
    print("opening url: " + url + " ...")
    driver.get(url)
    print("url opened!")

    wait = WebDriverWait(driver, 30)
    
    try:
        elem = wait.until(
        EC.presence_of_element_located((By.XPATH, '//*[@class="winner"]/a[2]'))
        )
    finally:
        winnerElements = driver.find_elements(By.XPATH, '//*[@class="winner"]/a[2]')   
    print("game box score links gotten!")
    print("converting WebElements to links...")
    winnerLinks = []
    for i in range( len(winnerElements) ):
        winnerLinks.append(winnerElements[i].get_attribute('href'))
    print("conversion done!")
    
    #print("printing links to check:")
    #print(winnerLinks[len(winnerLinks)-1].get_attribute('href'))
    
    print("quitting driver")
    driver.quit()
    print("FINISH 'getTourneyGameLinks'")
    print()
    
    return winnerLinks
```

```python
#PRE: given game url and headlessMode bool 
#POST: returns array of game stats for both teams formatted:
#[[team1 name, team1 basic stats HTML table, team1 advanced stats],
#[team2 name, team2 basic stats HTML table, team2 advanced stats]]]
def getGameBoxScores(url, headlessMode):
    print("-> START 'getGameBoxScores'")

    DRIVER_PATH = '/path/to/chromedriver'

    print("headless mode: " + str(headlessMode))

    options = Options()
    #configure non/headless mode:
    options.headless = headlessMode
    #configure window size:
    #options.add_argument("--window-size=1920,1200")
    driver = webdriver.Chrome(options=options, executable_path=DRIVER_PATH)

    #open given url
    print("opening url: " + url + " ...")
    driver.get(url)
    print("url opened!")

    wait = WebDriverWait(driver, 30)

    #//*[@id="boxes"]/div[2]/div/h2
    
    #getting first team name
    try:
        elem = wait.until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="boxes"]/div[2]/div/h2'))
        )
    finally:
        team1 = driver.find_element(By.XPATH, '//*[@id="boxes"]/div[2]/div/h2').text   
    
    #get basic stats data for team1
    try:
        elem = wait.until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="boxes"]/div[2]/div[2]/table'))
        )
    finally:
        basicStats1HTML = driver.find_element(By.XPATH, '//*[@id="boxes"]/div[2]/div[2]/table').get_attribute('innerHTML') 
    
    #get advanced stats data for team 1
    try:
        elem = wait.until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="boxes"]/div[3]/div[2]/table'))
        )
    finally:
        advStats1HTML = driver.find_element(By.XPATH, '//*[@id="boxes"]/div[3]/div[2]/table').get_attribute('innerHTML')
    #team 1 data in array
    team1data = [team1, basicStats1HTML, advStats1HTML]
    
    
    #getting second team name
    try:
        elem = wait.until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="boxes"]/div[5]/div/h2'))
        )
    finally:
        team2 = driver.find_element(By.XPATH, '//*[@id="boxes"]/div[5]/div/h2').text 
    
    #get basic stats data for team 2
    try:
        elem = wait.until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="boxes"]/div[5]/div[2]/table'))
        )
    finally:
        basicStats2HTML = driver.find_element(By.XPATH, '//*[@id="boxes"]/div[5]/div[2]/table').get_attribute('innerHTML') 
    
    #get advanced stats data for team 2
    try:
        elem = wait.until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="boxes"]/div[6]/div[2]/table'))
        )
    finally:
        advStats2HTML = driver.find_element(By.XPATH, '//*[@id="boxes"]/div[6]/div[2]/table').get_attribute('innerHTML')    
    #team 2 data in array 
    team2data = [team2, basicStats2HTML, advStats2HTML]
    
    gameStats = [team1data, team2data]
    
    print("quitting driver")
    driver.quit()
    print("FINISH 'getGameBoxScores'")
    print()

    return gameStats
    
```

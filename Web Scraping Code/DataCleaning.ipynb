```python
#PRE: path is the file path to the file to be cleaned, fileName is the file's name (without extension), saveLoc is the 
#     location the cleaned data is to be saved to
#POST: file at given location has unnecessary columns dropped and is reformatted then saved to given save location
def cleanData(path, fileName, saveLoc):
    #make sting of location of file
    loc = path + "/" + fileName + '.csv'
    
    #remove unneeded header row
    df = pd.read_csv(loc)
    df.columns = df.iloc[0]
    df = df[1:]
    
    #save for ease of later manipulation
    testFolder =  "C:/Users/danna/Documents/GitHub/WBBTournamentPredictions/Web Scraping Code/GatheredData/testingClean"
    dataSave(df, testFolder, "temp", ".csv")
    
    #reading in temp from above
    df2 = pd.read_csv(testFolder+"/temp.csv")
    
    #removing unneeded columns
    cols2Drop = ['MP', 'FG%', '2P%', '3P%', 'FT%', 'TRB']

    for col in df2.columns:
        if('Unnamed:' in col):
            df2.drop(col, axis=1, inplace=True)
    for col in cols2Drop:
        df2.drop(col, axis=1, inplace=True)
    
    #only keep starter's and total stat rows
    rows2Drop = []
    for i in range(5, len(df2.index)-1):
        rows2Drop.append(i)
    df2.drop(rows2Drop, inplace=True)

    #add 2 extra data columns
    ft2fga = []
    ast2tov = []
    for index, row in df2.iterrows():
        if (int(row['FGA']) == 0):
            ft2fga.append(0)
        else:
            ft2fga.append( int(row['FT']) / int(row['FGA']) )
        if( int(row['TOV']) == 0 ):
            ast2tov.append(0)
        else:
            ast2tov.append( int(row['AST']) / int(row['TOV']) )
    df2['FT/FGA'] = ft2fga
    df2['AST/TOV'] = ast2tov
    
    dataSave(df2, saveLoc, fileName ,'.csv')
    print("Saved", fileName, "to", saveLoc)
```

```python
#Cleans data in specified file locations

# import OS module
import os
 
# Get the list of all files and directories
path = "C:/Users/danna/Documents/GitHub/WBBTournamentPredictions/Web Scraping Code/GatheredData/3.14.23 230pm Scrape"
team1File = "Team 1 Basic BS"
team2File = "Team 2 Basic BS" 
ext = ".csv"
yr_list = os.listdir(path)
print(yr_list)
gamesLists = []
for yr in yr_list:
    path2 = path + "/" + yr
    gamesLists.append(os.listdir(path2))

yrIndex = 0
gameIndex = 1
#get location of team 1 file
loc = path + "/" + yr_list[yrIndex] + "/" + gamesLists[yrIndex][gameIndex] + "/" + team1File + ext
#read file in as a dataframe
df = pd.read_csv(loc)

filePath = path + "/" + yr_list[yrIndex] + "/" + gamesLists[yrIndex][gameIndex]

#establishing base location to have cleaned data saved to
saveFolder = "C:/Users/danna/Documents/GitHub/WBBTournamentPredictions/Web Scraping Code/GatheredData/CleanData"

for yr in range( len(yr_list) ):
    #make folder for season
    sznPath = saveFolder + '/' + yr_list[yr]
    if not os.path.isdir(sznPath):
        os.makedirs(sznPath)  
    for game in range( len( gamesLists[yr] ) ):
        #make folder for game's data to be saved at 
        gamePath = sznPath + '/' + gamesLists[yr][game]
        if not os.path.isdir(gamePath):
            os.makedirs(gamePath)
        #make paths and clean data for both teams
        gameLoc =  yr_list[yr] + "/" + gamesLists[yr][game]
        filePath = path + "/" + gameLoc
        saveLoc = saveFolder + "/" + gameLoc
        cleanData(filePath, team1File, saveLoc)
        cleanData(filePath, team2File, saveLoc)
```
